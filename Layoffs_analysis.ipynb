```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df= pd.read_csv('layoffs.csv')
```

```python
df.head()
```

```python
dfa=df.drop(['date','stage','funds_raised','percentage_laid_off'],axis=1)
```

```python
dfa.total_laid_off = dfa.total_laid_off.fillna(1)
dfa.total_laid_off = dfa.total_laid_off.astype(int)
dfa.industry = dfa.industry.fillna('NA')
```

```python
dfa.shape
```

```python
dfa.isna().sum()
```

```python
dfa.groupby('country')['total_laid_off'].sum().sort_values(ascending=False).head().plot(ylabel=" ", figsize=(8,8), kind='pie', stacked=True,colormap='Accent')

```

```python
dfa.groupby('industry')['total_laid_off'].sum().sort_values(ascending= False).head().plot(title="2022 laysoffs", ylabel="2022 layoffs as per industry", figsize=(12,8),kind='bar', stacked= True,colormap='Set2')
```

```python
dfa.groupby('industry')['total_laid_off'].sum().sort_values(ascending=False).head().plot(title ="2022 layoffs by industry world wide" ,ylabel="No of layoffs", figsize=(12,6), kind='bar', stacked=True, colormap='Set2')



```

```python
dfa.groupby('country')['total_laid_off'].sum().sort_values(ascending=False).head().plot(title="",ylabel="",figsize=(12,8),kind='bar',stacked=True, colormap='Set2')
```

```python
dfa.groupby('location')['total_laid_off'].sum().sort_values(ascending=False).head().plot(ylabel='', figsize=(8,8),kind='pie', stacked=True,colormap="Set3")
```

```python
dfa.groupby('location')['total_laid_off'].sum().sort_values(ascending=False).head().plot(ylabel="", figsize=(12,8),kind='bar',stacked=False, colormap='Set2')
```
